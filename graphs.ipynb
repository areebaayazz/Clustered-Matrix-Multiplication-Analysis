```python
import matplotlib.pyplot as plt
import numpy as np

# Data
clusters = [1, 2, 3]
process_2 = [0.000446, 0.000605, 0.000314]
process_4 = [0.000457, 0.000569, 0.000422]
process_6 = [0.000731, 0.000555, 0.000469]
process_8 = [0.000569, 0.000613, 0.007283]

bar_width = 0.2  # Adjust as needed

# Bar positions
bar_positions_2 = np.arange(len(clusters))
bar_positions_4 = [pos + bar_width for pos in bar_positions_2]
bar_positions_6 = [pos + bar_width for pos in bar_positions_4]
bar_positions_8 = [pos + bar_width for pos in bar_positions_6]

# Create bar plots
plt.bar(bar_positions_2, process_2, width=bar_width, label='Processs - 2')
plt.bar(bar_positions_4, process_4, width=bar_width, label='Processs - 4')
plt.bar(bar_positions_6, process_6, width=bar_width, label='Processs - 6')
plt.bar(bar_positions_8, process_8, width=bar_width, label='Processs - 8')

# Add labels and title
plt.xlabel('Cluster Machines')
plt.ylabel('Blocking Call Process')
plt.title('Blocking Call Process vs. Cluster Machines')
plt.xticks([pos + 1.5 * bar_width for pos in bar_positions_2], clusters)
plt.legend()

# Show the plot
plt.show()
```
